```python
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
n, p = 10000,  .5
b = np. random.binomial(n, p, 100000)
sns.distplot(b)
```

```python
# 3. Last let's try the same process with the poisson distribution (np.random.poisson)
#    poisson distribution:  discrete probability distribution that expresses the probability of a given number 
#    of events occurring in a fixed interval of time or space if these events occur with a known constant mean 
#    rate and independently of the time since the last event.

# n, p, s = 2, 1/6 ,  1000
p = np.random.poisson(100, 1000)
sns.distplot(p)
```
